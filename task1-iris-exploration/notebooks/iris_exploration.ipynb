## Objective
This notebook demonstrates how to load, inspect, and visualize a dataset to understand data trends, feature distributions, and relationships between variables using the classic Iris dataset.

## 1. Import Required Libraries

```python
# Import pandas for data manipulation and analysis (DataFrame, CSV handling)
import pandas as pd

# Import numpy for numerical operations on arrays and matrices
import numpy as np

# Import matplotlib.pyplot for creating static, interactive visualizations
import matplotlib.pyplot as plt

# Import seaborn for statistical data visualization (built on matplotlib)
import seaborn as sns

# Configure seaborn style: 'whitegrid' adds a white background with gridlines
sns.set_style('whitegrid')

# Set default figure size (width=12 inches, height=8 inches) for all plots
plt.rcParams['figure.figsize'] = (12, 8)

# Increase default font size for better readability in plots
plt.rcParams['font.size'] = 12

# Confirmation message to verify all imports succeeded
print('Libraries imported successfully')
```

## 2. Load the Dataset

```python
# Load the Iris dataset using seaborn's built-in dataset loader
# The dataset contains 150 samples of iris flowers with 4 features and 3 species
iris = sns.load_dataset('iris')

# Display confirmation message after successful data loading
print('Dataset loaded successfully!')
```

## 3. Inspect the Data

We will examine the shape, column names, first few rows, data types, and summary statistics.

```python
# Get the dimensions of the dataset: rows (samples) and columns (features)
# shape returns a tuple: (number_of_rows, number_of_columns)
print(f"Dataset Shape: {iris.shape}")
print(f"Rows: {iris.shape[0]}, Columns: {iris.shape[1]}\n")
```

```python
# List all column names in the dataset
# columns returns an Index object with all column labels
print("Column Names:")
for i, col in enumerate(iris.columns, 1):
    print(f"  {i}. {col}")
```

```python
# Display the first 5 rows to get a quick preview of the data
# head() shows the top n rows (default=5) to understand data structure
print("First 5 rows of the dataset:")
iris.head()
```

```python
# Get concise summary of the DataFrame: data types, non-null counts, memory usage
# info() is useful for checking missing values and data type correctness
print("Dataset Info:")
iris.info()
```

```python
# Generate descriptive statistics for numerical columns
# describe() computes: count, mean, std, min, 25%, 50%, 75%, max
# This helps understand the central tendency and spread of each feature
print("Summary Statistics:")
iris.describe()
```

## 4. Data Visualization

### 4.1 Scatter Plots - Relationships Between Features

Scatter plots help us visualize how features relate to each other and how species are separated.

```python
# Create a scatter plot to examine the relationship between sepal length and sepal width
# Different colors and markers represent different species for visual distinction
plt.figure(figsize=(10, 6))  # Create a new figure with specific size
sns.scatterplot(
    data=iris,              # DataFrame containing the data
    x='sepal_length',       # Column for x-axis
    y='sepal_width',        # Column for y-axis
    hue='species',          # Color points by species for visual grouping
    style='species',        # Use different marker styles per species
    s=100                   # Set marker size for better visibility
)
plt.title('Sepal Length vs Sepal Width by Species', fontsize=14, fontweight='bold')
plt.xlabel('Sepal Length (cm)')   # Label x-axis with units
plt.ylabel('Sepal Width (cm)')    # Label y-axis with units
plt.legend(title='Species')      # Add legend to identify species colors
plt.tight_layout()               # Adjust spacing to prevent label clipping
plt.show()                        # Display the plot
```

```python
# Scatter plot for petal length vs petal width
# Petal features are expected to show better species separation than sepal features
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', style='species', s=100)
plt.title('Petal Length vs Petal Width by Species', fontsize=14, fontweight='bold')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
# Create a pair plot: matrix of scatter plots for all feature combinations
# Diagonals show histograms (KDE) for each feature's distribution
# This gives a comprehensive view of all pairwise relationships at once
plt.figure(figsize=(10, 10))
sns.pairplot(
    iris,                   # Dataset to plot
    hue='species',          # Color by species
    diag_kind='hist',       # Show histograms on diagonal (distribution of single feature)
    markers=['o', 's', 'D'] # Circle, square, diamond markers per species
)
plt.suptitle('Pair Plot of Iris Features by Species', y=1.02, fontsize=14, fontweight='bold')
plt.show()
```

### 4.2 Histograms - Feature Distributions

Histograms show how each feature's values are distributed across the dataset.

```python
# Create histograms for all numeric features to understand their distribution
# KDE (Kernel Density Estimate) overlay shows the probability density curve
fig, axes = plt.subplots(2, 2, figsize=(14, 10))  # 2x2 grid of subplots
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
colors = ['skyblue', 'lightgreen', 'salmon', 'plum']  # Distinct colors per feature

# Loop through features and create a histogram for each
for ax, feature, color in zip(axes.flatten(), features, colors):
    sns.histplot(
        data=iris,              # Dataset
        x=feature,              # Feature to plot
        kde=True,               # Overlay KDE curve for smoothed distribution
        color=color,            # Fill color
        bins=20,                # Number of bins (intervals)
        ax=ax                   # Which subplot to draw on
    )
    # Format the feature name for the title (replace underscores, capitalize)
    formatted_name = feature.replace('_', ' ').title()
    ax.set_title(f'Distribution of {formatted_name}', fontweight='bold')
    ax.set_xlabel(f'{formatted_name} (cm)')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Histograms split by species to compare distributions across categories
# This reveals how different species have different feature value ranges
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for ax, feature in zip(axes.flatten(), features):
    # Plot a separate histogram for each species on the same axes
    for species in iris['species'].unique():
        subset = iris[iris['species'] == species]  # Filter data for this species
        sns.histplot(
            subset[feature],                        # Feature values for this species
            kde=True,                               # Add density curve
            label=species.capitalize(),             # Legend label
            alpha=0.5,                              # Transparency for overlapping bars
            bins=15,                                # Number of bins
            ax=ax
        )
    ax.set_title(f'{feature.replace("_", " ").title()} by Species', fontweight='bold')
    ax.set_xlabel(f'{feature.replace("_", " ").title()} (cm)')
    ax.set_ylabel('Frequency')
    ax.legend()  # Show species legend

plt.tight_layout()
plt.show()
```

### 4.3 Box Plots - Outlier Detection

Box plots help identify outliers and show the distribution of data through quartiles.

```python
# Create box plots to visualize feature distributions and detect outliers per species
# Box plot elements: median (line), IQR (box, Q1-Q3), whiskers (1.5*IQR), outliers (dots)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for ax, feature in zip(axes.flatten(), features):
    sns.boxplot(
        data=iris,           # Dataset
        x='species',         # Categories on x-axis
        y=feature,           # Numerical feature on y-axis
        palette='Set2',      # Color palette for the boxes
        ax=ax
    )
    ax.set_title(f'Box Plot of {feature.replace("_", " ").title()} by Species', fontweight='bold')
    ax.set_xlabel('Species')
    ax.set_ylabel(f'{feature.replace("_", " ").title()} (cm)')

plt.tight_layout()
plt.show()
```

```python
# Create a combined box plot for all features using melted (long-format) data
# Melting transforms columns into rows for easier multi-feature plotting
plt.figure(figsize=(12, 6))

# Convert from wide to long format: species stays as id, features become variable-value pairs
iris_melted = iris.melt(id_vars='species', var_name='feature', value_name='value')

sns.boxplot(
    data=iris_melted,      # Long-format data
    x='feature',           # Feature names on x-axis
    y='value',             # Feature values on y-axis
    hue='species',         # Color by species
    palette='Set2'
)
plt.title('Distribution of All Iris Features by Species', fontsize=14, fontweight='bold')
plt.xlabel('Feature')
plt.ylabel('Value (cm)')
plt.xticks(rotation=45)  # Rotate x-labels for readability
plt.tight_layout()
plt.show()
```

### 4.4 Correlation Heatmap

A correlation heatmap shows how strongly features are related to each other.

```python
# Compute and visualize the correlation matrix for numerical features
# Correlation ranges from -1 (strong negative) to +1 (strong positive)
# 0 indicates no linear relationship between features

# Drop the 'species' column since it's categorical, then compute correlations
correlation_matrix = iris.drop('species', axis=1).corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation_matrix,              # The correlation values
    annot=True,                      # Show correlation coefficients in cells
    cmap='coolwarm',                 # Color scheme: blue=cold(neg), red=warm(pos)
    linewidths=1,                    # Width of lines between cells
    square=True,                     # Make cells square-shaped
    fmt='.3f',                       # Format coefficients to 3 decimal places
    cbar_kws={'shrink': 0.8}         # Shrink the color bar to 80% height
)
plt.title('Feature Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 5. Summary and Key Insights

### Dataset Overview
- **Total samples:** 150 (50 per species)
- **Features:** 4 numeric features (sepal length, sepal width, petal length, petal width)
- **Target:** 3 species (Setosa, Versicolor, Virginica)
- **Missing values:** None (clean dataset)

### Key Findings
1. **Petal features are highly correlated** (0.96 correlation between petal length and petal width) - these are the most useful features for classification.
2. **Setosa is clearly separable** from Versicolor and Virginica across all features, especially petal dimensions.
3. **Virginica has the largest petal dimensions** while Setosa has the smallest.
4. **Sepal width shows overlapping distributions** across all three species, making it less useful for discrimination.
5. **Minimal outliers** are present in the dataset - the data is clean with few extreme values.

### Visual Insights
- Scatter plots clearly show that **petal length and width** effectively separate all three species.
- Box plots confirm **no significant outliers** in any feature.
- Histograms show **petal features** have bimodal-like distributions due to species differences.
- The pair plot provides a comprehensive view of all feature relationships simultaneously.

### Conclusion
The Iris dataset demonstrates classic patterns in multivariate data analysis. The clear separation between species, particularly using petal measurements, makes it an excellent introductory dataset for understanding data visualization, feature relationships, and classification concepts.
